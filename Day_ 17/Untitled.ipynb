```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize']=(16,9)
plt.style.use('ggplot')
```

```python
# importing DataSet
data= pd.read_csv('xclara.csv')
print(data.shape)
data.head()
```

```python
#Getting the values and plotting it

f1=data['V1'].values
f2=data['V2'].values
#print(zip(f1,f2))

plt.scatter(f1,f2,c='blue',s=7)
```

```python
#Euclidean Distance Calculator

def dist(a,b,ax=1):
    return np.linalg.norm(a-b,axis=ax)
```

```python
#Number of clusters
X=np.array(list(zip(f1,f2)))
k=3
# X coordiantes of random centroids

c_x=np.random.randint(0,np.max(X)-20,size=k)

#Y coordinates of random centroids

c_y=np.random.randint(0,np.max(X)-20,size=k)

c= np.array(list(zip(c_x,c_y)),dtype=np.float32)
print(c)
```

```python
plt.scatter(f1,f2,c='#595959',s=7)
plt.scatter(c_x,c_y,marker='*',s=200,c='purple')
```

```python
#to store the value of centroids as it updates
c_old=np.zeros(c.shape)
#cluster lables(0,1,2)
clusters=np.zeros(len(X))
#error func. - Distance between new centroids an old centroids
error=dist(c,c_old,None)

#loop will run till the error becomes zero

while error!=0:
    for i in range(len(X)):
        distance=dist(X[i],c)
        cluster=np.argmin(distance)
        clusters[i]=cluster
# Storing the old centroids value    
    c_old=deepcopy(c)
# Finding the new centroids by taking the average value
    for i in range(k):
        points=[X[j] for j in range(len(X)) if clusters[j]==i]
        c[i]=np.mean(points,axis=0)
    error=dist(c,c_old,None)
```

```python
colors=['r','g','b','y','c','m']
fig,ax=plt.subplots()
for i in range(k):
    points=np.array([X[j] for j in range(len(X)) if clusters[j]==i])
    ax.scatter(points[:,0],points[:,1],s=7,c=colors[i])
ax.scatter(c[:,0],c[:,1],marker='*',s=200,c='#050505')
```

```python
from sklearn.cluster import KMeans

#Number of clusters

kmeans = KMeans(n_clusters=3)


# fitting the input data

kmeans = kmeans.fit(X)
#getting the cluster lablels
labels =kmeans.predict(X)
centeroid = kmeans.cluster_centers_
```

```python
#comparing with scikit-learn centroids
print(c)#from scratch
print(centeroid) #from sci-kit learn
```

```python
import numpy as np
import matplotlib.pyplot as  plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.rcParams['figure.figsize']=(16,9)

#Creating a sample dataset with 4 clusters
X,y=make_blobs(n_samples=800,n_features=3,centers=4)
```

```python
fig =plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0],X[:,1],X[:,2])
```

```python

kmeans = KMeans(n_clusters=4)


# fitting the input data

kmeans = kmeans.fit(X)
#getting the cluster lablels
labels =kmeans.predict(X)
C = kmeans.cluster_centers_
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0],X[:,1],X[:,2],c=y)
ax.scatter(C[:,0],C[:,1],C[:,2],marker='*',c='#050505',s=1000)
```
