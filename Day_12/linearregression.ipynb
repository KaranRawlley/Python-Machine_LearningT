```python
%matplotlib inline
import numpy as np
from pylab import *

pageSpeeds = np.random.normal(3.0,1.0,1000)

print(size(pageSpeeds))
purchaseAmt = 100 - (pageSpeeds + np.random.normal(0,0.1,1000))*3

scatter(pageSpeeds,purchaseAmt)
```

```python
from scipy import stats
slope,intercept,r_value,p_value,std_err = stats.linregress(pageSpeeds,purchaseAmt)
```

```python
r_value**2
```

```python
import matplotlib.pyplot as py

def predict(x):
    return slope * x + intercept
fitline = predict(pageSpeeds)

py.scatter(pageSpeeds,purchaseAmt)
py.plot(pageSpeeds,fitline,c='r')
py.show()
```

# Polynomial regression

```python
%matplotlib inline
from pylab import *
import numpy as np

np.random.seed(2)
pageSpeeds = np.random.normal(3.0,1.0,1000)
purchaseAmt=np.random.normal(50.0,10.0,1000)/pageSpeeds

scatter(pageSpeeds,purchaseAmt)
```

```python
x = np.array(pageSpeeds)
y = np.array(purchaseAmt)

p4 = np.poly1d(np.polyfit(x,y,4))
print(p4)
print(type(p4))
```

```python
import matplotlib as plt

xp = np.linspace(0,7,100) #generates an array (start,end,total no)

scatter(x,y)
plot(xp,p4(xp),c='r')
show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y,p4(x))
print(r2)
```
